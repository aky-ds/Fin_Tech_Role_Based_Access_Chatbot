```python
import pandas as pd
data=pd.read_csv('data\hr\hr_data.csv')
```

```python
data.head()
```

```python
text=data.apply(lambda x:' '.join(x.astype(str)),axis=1).tolist()
```

```python
md_file = "data/hr/hr.md"

with open(md_file, "w", encoding="utf-8") as f:
    for line in text:
        f.write(f"- {line}\n")

print(f"✅ Markdown saved to {md_file}")
```
